```python
#Kiểm tra các giá trị thiếu
if data.isnull().any().any():
    print("Có giá trị null")
else:
    print("Không có giá trị null")
```

```python
#Kiểm tra các giá trị trùng lặp
if data.duplicated().any():
    print("Có giá trị trùng lặp")
else:
    print("Không có giá trị trùng lặp")
```

```python
#Vẽ biểu đồ boxplot kiểm tra giá trị ngoại lai
col = ['Income','Age','Dependents','Desired_Savings_Percentage']
plt.figure(figsize = (10,6))
for col in col:
    sns.boxplot(data[col])
    plt.show()
```

Biểu đồ biểu diễn Income và Desired_Savings_Percent có xuất hiện giá trị ngoại lai. Tuy nhiên, các số liệu này vẫn được giữ lại để phân tích có thể chính xác nhất.

```python
#Biểu đồ số lượng người quan sát theo từng City Tier
sns.countplot(data=data, x="City_Tier",color = 'lightgreen')
plt.show()
```

```python
#Thêm cột nhóm tuổi
bins = [17, 25, 35, 45, 55, 65, 100]
labels = ['18-25', '26-35', '36-45', '46-55', '56-65', '65+']
data["Age_Group"] = pd.cut(data["Age"], bins=bins, labels=labels)
```

```python
data.head()
```

```python
data['Age_Group'].value_counts()
```

```python
#Biểu đồ tần suất theo từng nhóm tuổi
sns.countplot(data=data, x="Age_Group", color = 'lightblue')
plt.show()
```

```python
data2 = data2.drop(columns = 'Transaction Description')
```

```python
data2.head()
```

```python
data2['Date'] = pd.to_datetime(data2['Date'])
```

```python
if data2.duplicated().any():
  print("Có giá trị trùng lặp")
else:
  print("Không có giá trị trùng lặp")
```

```python
data2['Category'].value_counts()
```

```python
sns.boxplot(data2['Amount'])
```

Có thể thấy cột Amount có chứa các giá trị ngoại lai, tuy nhiên mục tiêu của bài toán là phân tích chi tiêu tài chính. Vậy nên các giá trị này vẫn nên giữ lại để kết quả phân tích có thể chính xác và trực quan nhất.  

```python
#Các nhóm mục đích chi tiêu
(data2.groupby('Type').get_group('Expense'))['Category'].value_counts()
```

- Chi tiêu phân bố khá đều giữa các danh mục, cho thấy người dùng có thói quen chi tiêu đều đặn ở nhiều lĩnh vực khác nhau.

- Rent (thuê nhà) chiếm tần suất cao nhất → đây có thể là chi phí cố định hàng tháng lớn nhất.

```python
(data2.groupby('Type').get_group('Income'))['Category'].value_counts()
```

- Thu nhập ít đa dạng hơn chi tiêu.

- Cả hai loại thu nhập có số lượng gần bằng nhau → có thể cho thấy nguồn thu nhập chính đến từ đầu tư và một phần khác từ các nguồn không xác định rõ (Other).

- Thiếu các danh mục như “Salary” → Có thể Salary bị nhầm lẫn và nằm trong Expense như hình đầu tiên.


```python
data2['Category_group'] = data2['Category'].apply(lambda x: 'Unnessary Expense' if x in ['Travel','Shopping','Entertainment'] else 'Necessary Expense')
```
